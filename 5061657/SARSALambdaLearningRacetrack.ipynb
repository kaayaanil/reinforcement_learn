```python
#State Generator
states=[]
for i in range(len(course)):
    for j in range(len(course[1])):
        for vu in range(5):
            for vr in range(5):
                states.append((i,j,vu,vr))
```

```python
#ALL POSIBLE ACTIONS

ACTION_SPACE = ((-1,-1),(-1,0),(-1,1),(0,-1),(0,1),(0,0),(1,-1),(1,0),(1,1))
```

```python
def is_terminal(state):
    return course[state[0],state[1]]==3
def is_wall(state):
    return course[state[0],state[1]]==1
def is_boundry(state):
    return state[0]>33 or state[1]>31
def is_start(state):
    return course[state[0],state[1]]==2
```

is_boundry specifices out of the track such as (0,0,0,0) or (34,34,0,0)

```python
def random_initial_state():
    x=np.random.choice(np.where(course==2)[0])
    y=np.random.choice(np.where(course==2)[1])
    return((x,y,0,0))
```

```python
def get_next_state(s,a):
    i,j = s[0],s[1]
    vu,vr = s[2],s[3]

    if vu + a[0] >4: #Speed upper limit 4
        vu = 4
    elif vu + a[0] <= 0: #Speed lower limit 0
        vu = 0
    else:
        vu += a[0]
    if vr + a[1] >4:
        vr = 4
    elif vr+a[1] <= 0:
        vr = 0
    else:
        vr += a[1]
    
    i = i - vu #Move for Y axis
    j = j + vr #Move for X axis

    return i,j,vu,vr
```

```python
rewards = {}
States = []
for s in states:
    if not is_terminal(s):
        for a in ACTION_SPACE:
            s2 = get_next_state(s,a)
            if not s2 in States:
                States.append(s2)
            if(s2[0]<0 or s2[1]<0):
                continue
            if is_boundry(s2) or s2[0]<0 or s2[1]<0 or is_wall(s2):
                rewards[(s,a,s2)] = -5
            elif is_terminal(s2):
                rewards[(s,a,s2)] = 10
            else:
                rewards[(s,a,s2)] = -1
```

```python
def max_dict(d):
    # returns the argmax (key) and max (value) from a dictionary
    # put this into a function since we are using it so often

    # find max val
    max_val = max(d.values())

    # find keys corresponding to max val
    max_keys = [key for key, val in d.items() if val == max_val]

    return random.sample(max_keys,1)[0], max_val #Get maxiumum action
```

```python
def epsilon_greedy(Q, s, eps=0.1):
    if np.random.random() < eps:
        return random.sample(ACTION_SPACE,1)[0] #Get random action
    else:
        a_opt = max_dict(Q[s])[0]
    return a_opt #Get maxiumum action
```

```python
def SARSA(LAMBDA,n_episodes):
    GAMMA = 0.9
    ALPHA = 0.1
    # initialize Q(s,a) and eligibility traces
    Q = {}
    E={}
    for s in States:
        Q[s] = {}
        E[s]={}
        for a in ACTION_SPACE:
            Q[s][a] = 0

    for it in range(n_episodes):
        for s in States:
            for a in ACTION_SPACE:
                E[s][a] = 0
        s=random_initial_state()
        a = epsilon_greedy(Q, s, eps=0.1)
        visited_state_action={}
        while not is_terminal(s):
            if (s,a) not in visited_state_action.keys():
                visited_state_action[(s,a)]=[]
            s2 = get_next_state(s, a)
            if is_boundry(s2) or s2[0]<0 or s2[1]<0 or is_wall(s2): #These conditions i will go back to the starts
                s2 = random_initial_state()
            r = rewards.get((s, a, s2),0)
            a2 = epsilon_greedy(Q, s2, eps=0.1)
            delta=r + GAMMA*Q[s2][a2] - Q[s][a]
            E[s][a]=E[s][a]+1
            for (s,a) in visited_state_action.keys():
#            for s in States:
#                for a in ACTION_SPACE:
                    Q[s][a] = Q[s][a] + ALPHA*delta*E[s][a]
                    E[s][a] = E[s][a]*GAMMA*LAMBDA
            # next state becomes current state
            s = s2
            a = a2
    policy = {}
    V = {}
    for s in States:
        a, max_q = max_dict(Q[s])
        policy[s] = a
        V[s] = max_q
    return V,policy
```

```python
LAMBDA = 0.95
V,policy=SARSA(LAMBDA=LAMBDA,n_episodes=1500)
```
